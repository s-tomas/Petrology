```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load the data
file_path = '/Users/tomassalazar/Desktop/Work/Krawcyzski Group/ESPM.Code/Petrology/Kamchatka/EPMA S.1 Analyses/EPMA RAW DATA/EPMA Session 1 -  Fvs.Cl.csv'
data = pd.read_csv(file_path)
data.columns = data.columns.str.strip()

Cl = data['Cl']
F = data['F']

# Create scatter plot with best-fit line
plt.figure(figsize=(10, 6))
colors = np.random.rand(len(Cl))
slope, intercept, r_value, p_value, std_err = stats.linregress(Cl, F)

plt.scatter(Cl, F, c=colors, cmap='viridis', label='Data Points')
plt.plot(Cl, slope * Cl + intercept, color='red', label=f'Best fit line\nSlope: {slope:.2f}, Intercept: {intercept:.2f}')
plt.title('Chlorine (Cl) vs. Fluorine (F)')
plt.xlabel('Cl (%)')
plt.ylabel('F (%)')
plt.legend()
plt.grid(True)

# Annotate data points
for j, txt in enumerate(range(len(Cl)), start=1):
    plt.annotate(txt, (Cl.iloc[j - 1], F.iloc[j - 1]), fontsize=10, ha='right', va='top', color='black')

plt.tight_layout()
plt.show()
```
