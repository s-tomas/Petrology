```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/tomassalazar/Desktop/Work/Krawcyzski Group/ESPM.Code/Petrology/Kamchatka/EPMA S.1 Analyses/EPMA RAW DATA/EPMA S.1 Oxide Data.csv'
data = pd.read_csv(file_path)
data.columns = data.columns.str.strip()

sum_percentages = data.sum(axis=1)

green_mask = (sum_percentages >= 98.5) & (sum_percentages <= 101.5)

green_samples = data[green_mask]

chlorine = green_samples['Cl']
fluorine = green_samples['F']  
oxygen = green_samples['O']    

sample_numbers = green_samples.index

plt.figure(figsize=(10, 6))

plt.bar(sample_numbers, chlorine, color='green', label='Chlorine')
plt.bar(sample_numbers, fluorine, color='orange', label='Fluorine', bottom=chlorine)
plt.bar(sample_numbers, oxygen, color='blue', label='Oxygen/Vacancies', bottom=chlorine + fluorine)

plt.title('Chlorine, Fluorine, and Oxygen/Vacancies in Green Zone Amphibole Samples')
plt.xlabel('Sample Number')
plt.ylabel('Amount (%)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
