```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/tomassalazar/Desktop/Work/Krawcyzski Group/ESPM.Code/Petrology/Kamchatka/EPMA S.1 Analyses/EPMA RAW DATA/Amphibole.csv'
data = pd.read_csv(file_path)
data.columns = data.columns.str.strip()

samples = data.iloc[13:20]


chlorine = samples['Cl']
fluorine = samples['F']  
oxygen = samples['O']     

sample_numbers = samples.index + 1  

plt.figure(figsize=(10, 6))

plt.plot(sample_numbers, chlorine, marker='o', color='green', label='Chlorine (Cl)')
plt.plot(sample_numbers, fluorine, marker='o', color='orange', label='Fluorine (F)')
plt.plot(sample_numbers, oxygen, marker='o', color='blue', label='Oxygen/Vacancies (O)')

plt.title('Comparison of Cl, F, and O Levels in Samples 15-21')
plt.xlabel('Sample Number')
plt.ylabel('Percentage Levels (%)')
plt.xticks(sample_numbers)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
