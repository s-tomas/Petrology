```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = '/Users/tomassalazar/Desktop/Work/Krawcyzski Group/ESPM.Code/Petrology/Kamchatka/EPMA S.1 Analyses/EPMA RAW DATA/EPMA S.1 Oxide Data.csv'
data = pd.read_csv(file_path)
data.columns = data.columns.str.strip()

Cl = data['Cl']
FeO = data['FeO']

in_range_mask = (FeO >= 9) & (FeO <= 13)

colors = ['green' if in_range else 'red' for in_range in in_range_mask]

outlier_indices = data[~in_range_mask].index.tolist()
print("Outlier Indices:", outlier_indices)

in_range_indices = data[in_range_mask].index.tolist()
print("New Index with In-Range Data:", in_range_indices)

plt.figure(figsize=(10, 6))
plt.scatter(Cl, FeO, c=colors, label='Data Points')
plt.axhline(y=9, color='blue', linestyle='--', label='9% FeO')
plt.axhline(y=13, color='blue', linestyle='--', label='13% FeO')
plt.title('Chlorine (Cl) vs. Iron Oxide (FeO) with Highlighted Range')
plt.xlabel('Cl (%)')
plt.ylabel('FeO (%)')
plt.legend()
plt.grid(True)

for j, txt in enumerate(range(len(Cl)), start=1):
    plt.annotate(txt, (Cl.iloc[j - 1], FeO.iloc[j - 1]), fontsize=10, ha='right', va='top', color='black')

plt.tight_layout()
plt.show()
```
